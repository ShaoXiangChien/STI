```python
#!pip install ckiptagger
#!pip install tensorflow
#!pip install gdown
#data_utils.download_data_gdown("./")   #download model from google drive
#參考: https://github.com/ckiplab/ckiptagger/blob/master/README.md
from pprint import pprint 
import pandas as pd
from ckiptagger import data_utils

df = pd.read_csv("萊豬 output.csv")

def ckiptagger_keyword_extraction(df):
    text = list(df.paragraph.astype(str)+df.title.astype(str))
    from ckiptagger import POS, NER, WS
    ws = WS("./data")
    pos = POS('./data')
    ner = NER('./data')

    ws_result = ws(text, sentence_segmentation = True, segment_delimiter_set = {",", "。", ":", "?", "!", ";"}) 
    pos_result = pos(ws_result)
    ner_result = ner(ws_result, pos_result)
    kwd_dict = {}
    ner_tag = ['PERSON', 'WORK_OF_ART', 'EVENT', 'FAC', 'LAW', 'PRODUCT', 'ORG'] 
    for news in ner_result:
        for entity in news:
            if entity[2] in ner_tag:
                if entity[3] not in kwd_dict: 
                    kwd_dict[entity[3]] = 1
                else:
                    kwd_dict[entity[3]] += 1
    return kwd_dict
    
```
